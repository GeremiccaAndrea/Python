```python
df[df.country == 'US'] #visualizza tutti i vini americani 
```

```python
#trova tutti i vini italiani che costano più di 200 euro 
df[(df['country'] == 'Italy') & (df['price'] > 200)] # in pandas and = & 
```

```python
#trovare i vini italiani e francesi 
df[(df['country'] == 'Italy') | (df['country'] == 'France')] # in pandas or si fa | 
```

```python
#trova i vini francesi,italiani e statunitensi
df[df['country'].isin(['Italy', 'France', 'US'])]
```

```python
df.loc[df['price'].notnull()] #stampa i vini dove abbiamo il prezzo
```

```python
df.loc[df['price'].isnull()] #stampa i vini dove non abbiamo il prezzo
```

```python
#visualizzare tutti i prezzi dei vini 
df['price'] 
```

```python
#trovare il prezzo minimo 
min(df['price'])
#oppure 
df['price'].min() #quello più corretto 
```

```python
#trova quello con il prezzo più alto 
df['price'].max()
```

```python
#trovare il nome del vino che costa di più 
df[df['price']== df['price'].max()]['title']
```

```python
#trovare il nome del vino che costa di meno , il suo prezzo e i suoi punti 
df[df['price']== df['price'].min()][['title', 'price', 'points']]
```

```python
#somma di tutti punti di tutti i vini 
df['points'].sum()
```

```python
#trovare il prezzo medio dei vini 
df['price'].mean()
```

```python
#nomi della casa prodruttrice e del vino che costano più della media 
df[df['price'] > df['price'].mean()][['winery', 'title']]
```

```python
#trovare il numero dei vini americani 
len(df[df['country'] == 'US'])
#oppure 
df[df['country'] == 'US']['country'].count()
```
